Ask the user for the total bill amount and the percentage they want to tip. Calculate the tip amount and the grand total, rounding the output to two decimal places.

```python
billamount = int(input("Enter the total bill amount :"))
tip = float(input("Enter the percentage u want to tip :"))
tipamount = (billamount * tip)/100
grandtotal = billamount + tipamount
print (grandtotal)
```
